```python
# Prepare the features and target
window = 1
true_positives = 0
false_positives = 0
true_negatives = 0
false_negatives = 0

for data in total:
    data = data[features]
    encoded_data = data.replace({'상대팀': category_encoding, '우리팀': category_encoding, 'H/A_N': category_encoding})
    for i in range(len(encoded_data) - window):
        elo_win_ratio = encoded_data.iloc[i]['우리팀_elo승률']
        actual_result = encoded_data.iloc[i + window]['결과']
        
        # 우리팀의 Elo 승률이 0.5보다 큰 경우를 1로 예측
        predicted_result = 1 if elo_win_ratio > 0.55 else 0
        
        # 실제 결과가 1인 경우
        if actual_result == 1:
            if predicted_result == 1:
                true_positives += 1
            else:
                false_negatives += 1
        # 실제 결과가 0인 경우
        else:
            if predicted_result == 1:
                false_positives += 1
            else:
                true_negatives += 1

# Precision, Recall, F1 Score 계산
precision = true_positives / (true_positives + false_positives)
recall = true_positives / (true_positives + false_negatives)
f1_score = 2 * (precision * recall) / (precision + recall)

# 결과 출력
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1_score)
```
